```python
#find the mean Glucose and highest Glucose values of this sample
# mean of GlucoseSample
meanGlucoseSample = st.mean(sample_df.loc[:,"Glucose"])
print(meanGlucoseSample)
```

```python
#max Glucose of the sample
maxGlucoseSample = max(sample_df.loc[:,"Glucose"])
print(maxGlucoseSample)
```

```python
#max Glucose of the population
maxGlucosePopulation = max(sample_df.loc[:,"Glucose"])
print(maxGlucosePopulation)
```

```python
# creating meanDictionary function to store the mean of population and glucose sample

class meanDictionary(dict):
    
    def __init__(self):
        self = dict()
    
    def add(self, key, value):
        self[key] = value
 
 
dictMean = meanDictionary()
 
dictMean.add('meanSample', meanGlucoseSample)
dictMean.add('meanPopulation', meanGlucosePopulation)
 
print(dictMean)


# moving the result file to results directory

result_file1 = open("C:/Users/prave/Documents/pds/assignment2/results/result_file1.txt","w")

# writing to file
result_file1.write(str(dictMean))

# close file
result_file1.close()
```

```python
# creating meanDictionary function to store the max of population and glucose sample

class maxDictionary(dict):
    
    def __init__(self):
        self = dict()
    
    def add(self, key, value):
        self[key] = value
 
 
dictMax = maxDictionary()
 
dictMax.add('maxSample', maxGlucoseSample)
dictMax.add('maxPopulation', maxGlucosePopulation)
 
print(dictMax)


# moving the result file to results directory

result_file2 = open("C:/Users/prave/Documents/pds/assignment2/results/result_file2.txt","w")

# writing to file
result_file2.write(str(dictMax))

# close file
result_file2.close()
```

```python
#Using bootstrap (replace= True), create 500 samples (of 150 observation each) from the
#population

bootstrap = sk.utils.resample(diabetes_df, replace=True, n_samples=500, random_state=150)

print(bootstrap)
```

```python
#find the average mean for BloodPressure

meanBMI = st.mean(bootstrap.loc[:,"BMI"])
print(meanBMI)
```

```python
#find the std BMI

stdBMI = st.stdev(bootstrap.loc[:,"BMI"])
print(stdBMI)
```

```python
#find the percentile BMI 

percentileBMI = np.percentile(bootstrap.loc[:,"BMI"], 98)
print(percentileBMI)
```

```python
#find the mean BMI of the population

meanBMI_pop = st.mean(diabetes_df.loc[:,"BMI"])
print(meanBMI_pop)
```

```python
#find the std BMI of the population

stdBMI_pop = st.stdev(diabetes_df.loc[:,"BMI"])
print(stdBMI_pop)
```

```python
#find the percentile BMI of the population

percentileBMI_pop = np.percentile(diabetes_df.loc[:,"BMI"], 98)
print(percentileBMI_pop)
```
